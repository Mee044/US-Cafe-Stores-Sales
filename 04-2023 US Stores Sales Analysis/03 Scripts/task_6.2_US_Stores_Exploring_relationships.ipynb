```python
### Creating Correlation Heatmap 
```

```python
# Create a subplot with matplotlib 2
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_cor.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

Black equates to 0, or no relationship between the variables, dark purple and black equate to negative correlations, and orange and light orange equate to positive correlations.

Sales
With “Profit”: A 0.8 coefficient indicates a strong positive correlation. 
With “Margin”: A 0.94 coefficient indicates a very strong positive correlation.  
With “Budget Profit”: A 0.82 coefficient indicates a strong correlation.
With "Budget Margin": 0.92 coefficient indicates a very strong positive correlation.
With "Total Expenses or Total Expenses to get the Product to Sell ": 0.2 coefficient indicates a weak correlation.
With "Marketing or Expenses in Marketing": 0.23 coefficient indicates a weak correlation.

Total Expenses
With "Marketing": A 0.97 coefficient indicates a very strong positive correlation.

Note: Budget Profit is expected Profit ($), Budget Margin is Expected Profit + Expected Total and Margin is Profit + Total Expenses ($) OR Sales - COGS ($);
